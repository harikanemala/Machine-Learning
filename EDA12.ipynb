```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset for classification
iris = load_iris()

# Create a DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Basic Data Exploration
print("Dataset Head:")
print(df.head())

print("\nSummary Statistics:")
print(df.describe())

print("\nClass Distribution:")
print(df['target'].value_counts())

# Visualizations

# Pairplot to visualize relationships between features, colored by target class
plt.figure(figsize=(10, 6))
sns.pairplot(df, hue='target')
plt.show()

# Correlation Matrix - Visualize correlation between features
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Boxplot to show feature distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, orient='h')
plt.title('Boxplot of Features')
plt.show()

# Feature Distributions - Histograms with KDE (Kernel Density Estimation) for each feature
plt.figure(figsize=(10, 6))
for feature in iris.feature_names:
    sns.histplot(df[feature], kde=True, label=feature)
plt.title('Feature Distributions')
plt.legend()
plt.show()


```
